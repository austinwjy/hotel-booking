```python
import pandas as pd
import numpy as np
```

```python
##bookings
bookings = pd.read_csv('bookings.csv', sep='|')
bookings
```

```python
#drop duplicates if any
bookings = bookings.drop_duplicates()
```

```python
#check missing values
bookings.isnull().sum()
```

```python
#check data types
bookings.info()
```

```python
# Convert date columns to datetime
bookings['start date'] = pd.to_datetime(bookings['start date'], errors='coerce')
bookings['end date'] = pd.to_datetime(bookings['end date'], errors='coerce')
bookings.info()
```

```python
#check outliers
bookings.describe()
```

```python
#check dates value
bookings.sort_values(by='start date')
```

```python
bookings.sort_values(by='start date')
```

```python
#correct dates
bookings['start date'] = bookings['start date'].apply(lambda x: x.replace(year=2016))
bookings['end date'] = bookings['end date'].apply(lambda x: x.replace(year=2016))
```

```python
#export csv
bookings.to_csv("bookings_cleaned.csv", index=False)
```
